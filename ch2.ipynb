## 變數類型的轉換

- 試著印出下列這句話

```python
km = 42.195
print("In order to become a marathon runner, you have to run " + km + " kilometers in a race.") # TypeError
```

## 用 `str()` 函數

```python
km = 42.195
print("In order to become a marathon runner, you have to run " + str(km) + " kilometers in a race.")
```

## 轉換變數類型的函數

- `float()`：轉換變數類型為 float
- `int()`：轉換變數類型為 int
- `complex()`：轉換變數類型為 complex
- `bool()`：轉換變數類型為 bool
- `str()`：轉換變數類型為 str

## 轉換變數類型的函數（2）

- 讓同學練習使用轉換變數類型的函數

```python
my_bool = True
print(type(my_bool))
print(float(my_bool))
print(int(my_bool))
print(complex(my_bool))
print(type(str(my_bool)))
print(type(bool(123)))
print(type(bool(2 + 5j)))
print(type(bool(0)))
```

- 有的時候要用兩次

```python
print(int(bool("True")))
```

## 布林判斷

- 讓同學練習布林判斷

|判斷運算符號|作用|
|--------|---|
|`==`|是否等於|
|`!=`|是否不等於|
|`>`|是否大於|
|`<`|是否小於|
|`>=`|是否大於等於|
|`<=`|是否小於等於|
|`in`|是否包含於|
|`not in`|是否不包含於|
|`and`|判斷兩側的條件是否都為 `True`|
|`or`|判斷兩側的條件是否至少有一側為 `True`|
|`not`|將布林的 `True` 或 `False` 反轉|

## 關於文字（2）

- 帶有格式的 `print`
    - `%s`：文字
    - `%f`：浮點數
    - `%i`：整數

````python
pi = 3.141591

print("圓周率是 %s" % pi)
print("圓周率是 %.2f" % pi)
print("圓周率整數部分是 %i" % pi)
print("%s是 %.4f" %("圓周率", pi))
````
