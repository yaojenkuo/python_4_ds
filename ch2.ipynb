## 變數類型的轉換

- 試著印出下列這句話

```python
km = 42.195
print("In order to become a marathon runner, you have to run " + km + " kilometers in a race.") # TypeError
```

## 用 `str()` 函數

```python
km = 42.195
print("In order to become a marathon runner, you have to run " + str(km) + " kilometers in a race.")
```

## 轉換變數類型的函數

- `float()`：轉換變數類型為 float
- `int()`：轉換變數類型為 int
- `complex()`：轉換變數類型為 complex
- `bool()`：轉換變數類型為 bool
- `str()`：轉換變數類型為 str

## 轉換變數類型的函數（2）

- 讓同學練習使用轉換變數類型的函數

```python
my_bool = True
print(type(my_bool))
print(float(my_bool))
print(int(my_bool))
print(complex(my_bool))
print(type(str(my_bool)))
print(type(bool(123)))
print(type(bool(2 + 5j)))
print(type(bool(0)))
```

- 有的時候要用兩次

```python
print(int(bool("True")))
```
